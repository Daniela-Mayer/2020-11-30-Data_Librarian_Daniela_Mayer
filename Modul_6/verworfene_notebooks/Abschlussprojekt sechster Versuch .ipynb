# 3. Beide Dateien an der PPNn mergen 

Vor dem Mergen muss bei dfCBS die PPN-Spalte in String-Werte umgewandelt werden.

```python
#Laut Konrad soll ich vor der Umwandung ein Re-Indexing machen, weil sonst Warnung käme. Hat bei mir auch
#ohne geklappt, aber testweise nun einmal das Reindexing durchführen:
dfCBSneu = dfCBSneu.reindex()
dfCBSneu['PPN'] = pd.Series(dfCBSneu['PPN'], dtype="string")
```

```python
#Nun Prüfen, wie viele Übereinstimmungen in den #PPN-Spalten der beiden Dateien vorkommen, wenn die PPN-
#Spalte in dfCBS in String umgewandelt wird:
set_1 = set(dfRepo.PPN)
set_2 = set(dfCBSneu.PPN.astype("str"))
len(set_2.intersection(set_1))
```

```python
#Erster Merge-Versuch:
Merge1 = dfRepo.merge(dfCBSneu, on='PPN', how='inner')
```

```python
Merge1.shape
```

```python
Merge1.head(15)
```

```python
#Als neue Excel-Tabelle speichern, um prüfen zu können:
Merge1.to_excel("Merge1_1.xlsx")
#Durch das Umwandeln in Srings hat es jetzt geklappt! 4660 datensätze wurden korrekt gemerged!
```

```python
#Versuch, mit left-join zu mergen:
Merge2 = dfRepo.merge(dfCBSneu, on='PPN', how='left', left_index=True)
```
