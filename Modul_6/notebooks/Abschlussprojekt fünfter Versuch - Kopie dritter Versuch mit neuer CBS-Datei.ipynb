#Aufgabe: 
Die Datei "AbschlussarbeitenReposit" (="Datei1") weist Abschlussarbeiten der HAW Hamburg nach und enthält Daten wie Autorin, Titel, Abstact, Jahr, ppn, etc.). Es fehlen dort die Namen der Betreuer*innen. Diese sollen ergänzt werden und können der Datei "AbschlussarbeitenCBS" (="Datei2") entnommen werden, welche z.B. Autor*in, Betreuer*innen, Titel, ppn enthält. Die Datensätze der beiden Dateien enthalten stimmen nur ungefähr überein, es gibt auch Datensätze die jeweils in nur einer der beiden Dateien auftauchen. D.h. ein einfaches Kopieren der Betreuer*innen-Spalte aus Datei2 und einfügen in Datei1 ist nicht möglich.
Der Identifikator, anhand dessen eine eindeutige Zuordnung der Datensätze möglich ist, ist die ppn (Pica Produktionsnummer).

```python
dfReposit = pd.read_excel("../data/AbschlussarbeitenReposit.xlsx")
dfCBS = pd.read_excel("../data/CBSneuRef.xlsx")
```
