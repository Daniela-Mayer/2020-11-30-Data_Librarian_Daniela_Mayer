```python
#Neu einlesen der neuen Excel-Datei führt dazu, dass die u.g. Operationen funktionieren. Warum geht es nicht
#weiter wie gehabt mit der Variable "dfCBSneu" ?? 
dfCBSneu = pd.read_excel("../data/CBSneu.xlsx")
```

```python
#Eventuelle Integer-Werte in Strings umwandeln, damit alle weiteren Operationen funktionieren
dfCBSneu['sonstPers'] = dfCBSneu['sonstPers'].astype("str")
```

```python
dfCBSneu.head(5)
```

```python
#Die Personen sollen in getrennten Spalten nachgewiesen werden: Der erste Ref. in einer Spalte namens "Ref", 
#alle Ko-Ref. in einer zweiten Spalte "RefID_plus_KoRef". 
#Für letzteres alles was bei "sonstPers" rechts des ersten Semikolons steht, in eine neue Spalte 
#"RefID_plus_KoRef" schreiben: 
dfCBSneu['RefID_plus_KoRef'] = dfCBSneu['sonstPers'].apply(lambda cell: ";".join(cell.split(";")[1:]))
```

```python
dfCBSneu.head(10)
```

```python
#Nun alles, was in der Spalte "sonstPers" links des ersten Semikolons steht, in eine neue Spalte 
#"Ref" schreiben:
dfCBSneu['Ref'] = dfCBSneu['sonstPers'].apply(lambda cell: ";".join(cell.split(";")[:1]))
```

```python
dfCBSneu.head(5)
```

```python
dfCBSneu.shape
```

```python
#Als neue Excel-Tabelle speichern, um prüfen zu können:
dfCBSneu.to_excel("../data/CBSneuRef.xlsx")
#Alle Spalten mit Personennamen sind korrekt befüllt!
```

```python
#Diese neue Excel-Datei einlesen
dfCBSneu = pd.read_excel("../data/CBSneuRef.xlsx")
```

```python
dfCBSneu.head(5)
```

```python
#Die beiden neu entstandenen Index-Spalten löschen
dfCBSneu = dfCBSneu.drop(['Unnamed: 0'], axis=1)
```

```python
dfCBSneu = dfCBSneu.drop(['Unnamed: 0.1'], axis=1)
```

```python
dfCBSneu.head(5)
```

```python
#In den Spalten "Ref" und RefID_plus_KoRef" soll alles entfernt werden, was nicht "Nachname, Vorname" ist. 
#D.h. alles was in eckigen Klammern steht, alle Zahlen und Sonderzeichen (außer dem Komma zwischen Nachname 
#und Vorname). Dazu eine Kopie der Spalte "Ref" anfertigen und mit str.extract das entfernen, was in der 
#Klammer mithilfe Regulärer Ausdrücke (RegEx) definiert wird 
#
#dfCBSneu['RefKopie'] = dfCBSneu['Ref']
#dfCBSneu['RefKopie'] = dfCBSneu['RefKopie'].str.extract('([\w \-\.]+,\s*[\w \-\.]+)')
#
#NEU: jetzt sollen auch die Leerzeichen vor dem Nachnamen entfernt werden, die man #nur in der CSV-Version 
#der Merge2-Datei sieht: 
#dfCBSneu['RefKopie'] = dfCBSneu['Ref']
#dfCBSneu['RefKopie'] = dfCBSneu['RefKopie'].str.extract('(\w[\w \-\.]*,\s*\w[\w \-\.]*[\w\.])')
#
#Weitere Verbesserung der RegEx (Ergänzung um $), um auch Namenszusätze "van" bzw. "von" mit auszugeben.  
#Außerdem benötigt der Punkt grunds. kein Escapen und der Bindestrich nicht, wenn er am Ende der [ ] steht:
dfCBSneu['RefKopie'] = dfCBSneu['Ref']
dfCBSneu['RefKopie'] = dfCBSneu['RefKopie'].str.extract('(\w[\w .-]*,\s*\w[\w .$-]*[\w\.])')
```

```python
dfCBSneu.head(10)
```

```python
#Funktioniert jetzt, auch bei Doppelnamen und mehreren Vornamen. 
```

```python
#Nun das gleiche für die Spalte RefID_plus_KoRef" 
dfCBSneu['RefID_plus_KoRefKopie'] = dfCBSneu['RefID_plus_KoRef']
dfCBSneu['RefID_plus_KoRefKopie'] = dfCBSneu['RefID_plus_KoRefKopie'].str.extract('(\w[\w .-]*,\s*\w[\w .$-]*[\w\.])')
```

```python
dfCBSneu.head()
```

```python
#Funktioniert jetzt, auch bei Doppelnamen und mehrere Vornamen. Aber es werden auch mögliche zweite Ko-Referenten 
#entfernt. 
```

```python
#Als excel-Datei speichern und anschauen: 
dfCBSneu.to_excel("../data/CBSneuBereinigt.xlsx")
```

```python
#Bei den Namenszusätzen (von oder van) wird jetzt auch "$c" ausgegeben. Dies soll im ganzen Dataframe entfernt
#werden. Mit der Methode replace() das "$c" durch ein Leerzeichen ersetzen. 
#dfCBSneu = dfCBSneu.replace("$c", " ") #klappt nicht, keine Änderung
#dfCBSneu = dfCBSneu.replace("[$c]", " ") #dadurch werden auch alle einzelnen "c" gelöscht
dfCBSneu = dfCBSneu.replace(["$c"], " ")
```

```python
#Als excel-Datei speichern und anschauen: 
dfCBSneu.to_excel("../data/CBSneuBereinigt2.xlsx")
```

# 3. Beide Dateien an der PPNn mergen 

Vor dem Mergen muss bei dfCBS die PPN-Spalte in String-Werte umgewandelt werden.

```python
#Laut Konrad soll ich vor der Umwandung ein Re-Indexing machen, weil sonst Warnung käme. Hat bei mir auch
#ohne geklappt, aber testweise nun einmal das Reindexing durchführen:
dfCBSneu = dfCBSneu.reindex()
dfCBSneu['PPN'] = pd.Series(dfCBSneu['PPN'], dtype="string")
```

```python
#Nun Prüfen, wie viele Übereinstimmungen in den #PPN-Spalten der beiden Dateien vorkommen, wenn die PPN-
#Spalte in dfCBS in String umgewandelt wird:
set_1 = set(dfRepo.PPN)
set_2 = set(dfCBSneu.PPN.astype("str"))
len(set_2.intersection(set_1))
```

```python
#Erster Merge-Versuch:
Merge1 = dfRepo.merge(dfCBSneu, on='PPN', how='inner')
```

```python
Merge1.shape
```

```python
Merge1.head(15)
```

```python
#Als neue Excel-Tabelle speichern, um prüfen zu können:
Merge1.to_excel("../data/Merge1_1.xlsx")
#Durch das Umwandeln in Srings hat es jetzt geklappt! 4660 datensätze wurden korrekt gemerged!
```

```python
#Alternativ zu Merge1 nun mit left-join:
Merge2 = dfRepo.merge(dfCBSneu, on='PPN', how='left', left_index=True)
```

```python
#Den Inhalt der Spalte "RefKopie" in die Spalte "dc.contributor.advisor" kopieren sowie Inhalt der Spalte 
#"RefID_plus_KoRefKopie" in die Spalte "tuhh.contributor.referee" kopieren 
Merge2['dc.contributor.advisor'] = Merge2['RefKopie']
Merge2['tuhh.contributor.referee'] = Merge2['RefID_plus_KoRefKopie']
```
