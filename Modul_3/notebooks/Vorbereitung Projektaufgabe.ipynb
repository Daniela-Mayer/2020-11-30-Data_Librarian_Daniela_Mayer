#Erstmal arbeiten mit einer Kopie von "Erstellung einer neuen Variable Membership Duration"
#falls ich die Spalte "Membership Duration" für die Projektaufgabe verwenden will

```python
df = pd.read_csv("../data/Library_Usage.csv")
df.head(10)
```

```python
#Die Spalte "Circulation Active Year" ist als Text gespeichert. Sie soll in ein
#numerisches Format konvertiert werden mit der Methode pd.to_numeric()
#Die Variable 'Circulation Active Year' soll mit den neuen Werten überschrieben werden
df['Circulation Active Year'] = pd.to_numeric(df['Circulation Active Year'], errors='coerce')
df['Circulation Active Year'] 
```

```python
df.head(10)
#Erneute Anzeige der Tabelle zeigt, dass die o.g. Werte in der Spalte umgewandelt wurden. 
```

```python
#Die Spalte "Circulation Active Month" enthält die Monatsnamen als Text gespeichert. 
#Diese sollen in ein numerisches Format konvertiert werden. 
#Zuerst die Spalte in ein Datumsformat konvertieren mit der Funktion pd.to_datetime
#Die Variable soll mit den neuen Werten überschrieben werden
df['Circulation Active Month'] = pd.to_datetime(df['Circulation Active Month'], errors='coerce', format="%B")
df['Circulation Active Month']
```

```python
#Nun aus diesen Zahlen den Monat extrahieren
df['Circulation Active Month'] = df['Circulation Active Month'].dt.month
df['Circulation Active Month']
```

```python
#Nun kann die aktive Mitgliedschaftsdauer in Monaten berechnet werden
#Das Ergebnis soll in eine neue Spalte "Membership Duration" geschrieben werden
df['Membership Duration'] = (df['Circulation Active Year'] - df['Year Patron Registered'])*12 + df['Circulation Active Month']
```

```python
#Es gibt fehlende Werte in dieser Spalte. Diese sollen mit 0 (null) ersetzt werden.
#D.h. alle NaN-Werte sollen mit 0 ersetzt werden. 
df['Membership Duration'] = df['Membership Duration'].fillna(0)
df['Membership Duration']
```

```python
#Tabelle enthält nun eine neue Spalte "Membership Duration"
df.head(10)
```

```python
#Mit einem bool'schen Zeilenfilter kann man ausgesuchte Werte zweier Variablen filtern und verbinden
#Z.B. hier: alle Senioren, die eine Mailadresse als Kontakt angegeben haben:
row_filter = (df['Patron Type Definition'] == 'SENIOR') & (df['Notice Preference Definition'] == 'email')
df.loc[row_filter]
```
