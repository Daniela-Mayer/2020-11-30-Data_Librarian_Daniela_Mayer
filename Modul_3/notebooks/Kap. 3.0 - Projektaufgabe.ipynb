#NEUER ANSATZ (aus der Musterlösung): Beispielfrage 3: "Wie alt sind die Nutzer, die per Mail informiert werden möchten, durchschnittlich im Vergleich zu Nutzern, die per Post informiert werden möchten?". Eventueller Lösungsweg: 
1..Durchschnittliches Alter der Gruppe von Nutzern die email angegeben haben
2..Durchschnittliches Alter der Gruppe von Nutzern, die print angegeben haben
3..Vergleich von 1. und 2. 

```python
#Spalte mit der Benachrichtigungsart anschauen: Es handelt sich dabei um ein nominales Merkmal 
#(keine logische Reihenfolge möglich):
df['Notice Preference Definition'].value_counts()
```

```python
#Age Range ist ein ordinales Merkmal. Um das nominale Merkmal (N.P.D.) mit dem ordinalen Merkmal 
#Age Range verbinden zu können, erstellen wir einen Plot. Dafür erst einmal eine neue Hilfsvariable 
#definieren, die nur die Merkmale "print" und "email" enthält:
benchrichtigung=df.loc[
    (df)['Notice Preference Definition'] == 'email')
    (df)['Notice Preference Definition'] == 'print')
]
```

```python
sns.catplot(x='Age Range', kind='count', hue='Notice Preference Definition', 
            data=benachrichtigung, aspect=3, order=["0 to 9 years","10 to 19 years",
                                                    "20 to 24 years","25 to 34 years",
                                                    "35 to 44 years","45 to 54 years", "55 to 59 years",
                                                    "60 to 64 years","65 to 74 years","75 years and over"])
```

```python
#Um die Zahlen besser verstehen zu können, hilft eine Kreuztabelle.
#Hier steht bei der Summe der Nutzer, die per Mail benachrichtigt werden möchten, eine andere Zahl als 
#die bei Frage 2. Grund: Bei einigen Nutzern, bei denen "email" steht, ist kein Eintrag in "Age Range"
#vorhanden. Diese Nutzer werden hier nicht ausgegeben. 
pd.crosstab(benachrichtigung[’Notice Preference Definition'], benachrichtigung['Age Range'], margins=True)
```

```python
#Nun das Ganze prozentual betrachten:
pd.crosstab
(benachrichtigung[’Notice Preference Definition'], 
                  benachrichtigung['Age Range'], 
                  margins=True, normalize=0)
```

```python
#Antwort also:
#Nutzer, die per Mail informiert werden möchten, sind
#ca. 9% in der Altersklasse 0 bis 9 Jahre, ca. 12% 10 bis 19 Jahre, ...
#Nutzer, die Per Post informiert werden möchten, sind
#ca. 13% in der Altersklasse 0 bis 9 Jahre, ca. 17% 10 bis 19 Jahre, ...
```
