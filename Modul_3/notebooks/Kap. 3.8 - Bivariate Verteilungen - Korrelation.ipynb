```python
#Für zwei metrische Variablen lässt sich der Zusammenhang über die sog. Kovarianz berechnen.
#Zusätzlich zur Kovarianz ist der Korrelationskoeffizient eine wichtige Kennzahl. 
#Der Korrelationskoeffizient ρx,y misst für zwei metrische Variablen x und y die Stärke des 
#linearen Zusammenhangs. 
#Der Korrelationskoeffizient stellt sozus. die standardisierte Kovarianz dar.
#Der Koeffizient kann Werte zwischen −1 (negativer Zusammenhang) und 1 (positiver Zusammenhang) annehmen. 
#Beispiel von zufällig generierten Variablen mit verschiedenen Korrelationskoeffizienten:
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.set()

n = 1000
cor = [-0.99, -0.7, -0.5, -0.2, 0, 0.2, 0.5, 0.7, 0.99]
dat = pd.DataFrame()

for c in cor:
    covmat = np.array((1, c, c, 1)).reshape((2, 2))
    X = np.random.multivariate_normal([0,0], covmat, size=n)
    X = np.hstack((X, np.ones(n).reshape((n,1))*c))
    dat = dat.append(pd.DataFrame(X, columns=["x", "y", "correlation"]))
    
g = sns.FacetGrid(col="correlation", data=dat, col_wrap=3)
g.map(sns.scatterplot, "x", "y")
g
```

```python
g.savefig('./correlation.png')
```
