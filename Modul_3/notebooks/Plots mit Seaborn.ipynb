```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set()
#Mit sample() eine zufällige Stichprobe erstellen
df = pd.read_csv("../data/Library_Usage.csv").sample(n=1000)
```

```python
#Einen Plot erzeugen mit der Funktion relplot() - ein Streudiagramm einer bivariaten 
#Verteilung mit metrischen Variablen. Jedes Wertepaar ist ein Punkt.
sns.relplot(x='Total Checkouts', y='Total Renewals', data=df)
```

```python
#Die Größe des Plots kann über die Argumente height und aspect konfiguriert werden
sns.relplot(x='Total Checkouts', y='Total Renewals', data=df, height=5, aspect=3)
```

```python
#Mit den Argumenten hue (Farbe), size und style weitere Anpassungen
sns.relplot(x='Total Checkouts', y='Total Renewals',
            hue='Provided Email Address', style='Outside of County',
            size='Year Patron Registered', data=df)
```

```python
#Plot als Bilddatei abspeichern mit savefig()
plt.savefig("../data/seaborn_tutorial.png", dpi=150)
```

```python
#Mit den Argumenten "row" und/oder "col" Plots anhand einer o.mehrerer Variablen in Form von Spalten 
#und/oder Zeilen anordnen:
sns.relplot(x='Total Checkouts', y='Total Renewals', 
                hue='Provided Email Address', style='Outside of County', 
                size='Year Patron Registered', data=df)
```

```python
#Mit dem Argument "palette" können Farbpaletten für den hue-Parameter ausgewählt werden.
#(Z.B. die Farben von ColorBrewer). Wichtig ist, die Anzahl der benötigten Farben mitzugeben.
sns.relplot(x='Total Checkouts', y='Total Renewals', 
            hue='Provided Email Address', 
            col='Outside of County',
            palette=sns.color_palette('Dark2', 2), 
            data=df)
```

#Weitere Beispiele für Plots mit Seaborn: Nominale und ordinale Variablen

```python
#Univariate Häufigkeitstabellen und Bivariate Kreuztabellen können mit Balkendiagrammen visualisiert werden
#Ab hier wieder den kompletten Datensatz verwenden, ohne Stichprobe wie oben
df = pd.read_csv("../data/Library_Usage.csv")
sns.catplot(y='Year Patron Registered', 
            data=df, kind='count', color="steelblue")
```

```python
sns.catplot(y='Age Range', hue='Provided Email Address', 
            data=df, kind='count')
```

```python
sns.catplot(x='Patron Type Definition', 
            data=df, kind='count', 
            col='Year Patron Registered', col_wrap=4)
```

#Weitere Beispiele für Plots mit Seaborn: Metrische Variablen

```python
#Univariate Verteilungen werden mit Histogrammen oder Kernel-Dichte-Schätzern visualisiert.
#Histogramm:
sns.distplot(df['Total Renewals'], kde=False)
```

```python
#Mit Kern-Dichte-Schätzer:
sns.distplot(df['Total Renewals'], kde=True)
```

#Kombination aus metrischen und nominalen/ordinalen Variablen

```python
#Swarmplot
sns.catplot(x='Year Patron Registered', y = 'Total Renewals',
            data=df, kind='swarm', color="steelblue", aspect=4)
```

```python
#Boxplot:
sns.catplot(col='Year Patron Registered', y = 'Total Renewals',
            data=df, kind='box', color="steelblue", aspect=4)
```
