#Daten aus bibsonomy (im json-Format) in Python bearbeiten

```python
import json
```

```python
url = "https://www.bibsonomy.org/search/universit%c3%a4t%20hamburg?format=json"
```

```python
import requests
```

```python
result = requests.get(url)
```

```python
data = result.json()
```

```python
type(data)
```

```python
print(data)
```

```python
print(json.dumps(data, indent=2))
```
