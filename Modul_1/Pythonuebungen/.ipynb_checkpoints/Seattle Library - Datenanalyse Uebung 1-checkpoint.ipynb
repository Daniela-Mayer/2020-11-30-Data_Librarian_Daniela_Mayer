```python
import urllib.request
import pandas as pd
```

```python
## Alle Medien die 150 mal ausgeliehen worden sind über die API beziehen und als CSV-Datei speichern
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?checkouts=150"
checkout150_file = "checkout150.csv"
urllib.request.urlretrieve(url, checkout150_file)
```

```python
## In einen DataFrame laden und analysieren
```

```python
checkouts_df = pd.read_csv(checkout150_file)
```
