```python
import urllib.request
import pandas as pd
```

## Checkout-Daten über die API beziehen und als CSV-Datei speichern

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2018&$limit=10000"
checkout_file = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file)
# where und limit sind Parameter. Was bewirkt where?
```

## In einen DataFrame laden und analysieren

```python
checkouts_df = pd.read_csv(checkout_file)
# Mit pd.read_csv() können Daten im csv-Format importiert und in einen DataFrame geladen werden
```

```python
checkouts_df
# Es werden nur die ersten und letzten fünf Einträge angezeigt. Wie kann ich mehr bzw. alle anzeigen lassen?
```

```python
# Es sieht so aus, als ob nur Entleihungen aus Januar ausgewiesen würden? 
# Die Funktion (?) unique zeigt an, welche einzelnen Werte in einem Datenset enthalten sind. 
# Also hier, welche Monate in der Spalte checkoutmonth auftauchen. 
# Stimmt diese Erkläuterung? Was macht der folg. Befehl genau?
# Ausgegeben werden best. Arrays - sind das die Werte, die in der o.g. Tabelle bei checkoutmonth vorkommen?
# Warum werden bei Konrads Tabelle nur Medien mit checkoutmonth 12 ausgegeben, obwohl wir exakt das gleiche abgefragt haben?
# Evtl. weil der Datenausschnitt (10000 Treffer) ein anderer ist?
checkouts_df["checkoutmonth"].unique()
```

```python
checkouts_df["materialtype"].value_counts()
# Die Funktion "value_counts()" gibt die Anzahl der einzelnen Werte aus der Spalte materialtype aus.
```

```python
%matplotlib inline
# Matplotlib ist eine "magische Funktion", d.h. von Pandas vordefiniert.
# Der Befehl "%matplotlib" richtet das JN für die Anzeige von Plots mit Matplotbib ein. 
# Die Ausgabe von Plotting-Befehlen wird dadurch direkt (inline) im Frontend, z.B. JN, angezeigt.
# D.h. auch die graphischen Ausgaben werden im JN gespeichert.
```

```python
checkouts_df["materialtype"].value_counts().plot(kind="bar")
# Die Liste oben soll in einem vertikalen Balkendiagramm ausgegeben werden. 
# plot = Grafik bzw. Diagramm, bar = Balken
```

```python
checkouts_df["materialtype"].value_counts().plot(kind="barh")
# Die Liste oben soll in einem horizontalen Balkendiagramm ausgegeben werden. 
```

```python
checkouts_df["checkouts"].hist()
# Die Ausleihen sollen in einem Histogramm (graf. Darstellung der Häufigkeitsverteilung) ausgegeben werden. 
# Was ist hier die Aussage der Grafik?? Macht diese Sinn?
```
