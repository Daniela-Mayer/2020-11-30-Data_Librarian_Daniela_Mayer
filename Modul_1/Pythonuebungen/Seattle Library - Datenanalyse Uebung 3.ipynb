```python
import urllib.request
import pandas as pd
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2020&$limit=5000"
checkout2020 = "checkout2020"
urllib.request.urlretrieve(url, checkout2020)
```

```python
checkouts_df = pd.read_csv(checkout2020)
```
